```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("final.csv")
```

```python
data.isnull().sum()
```

```python
data.fillna(data.mean(),inplace = True)
data.isnull().sum()
duplicate_rows = data[data.duplicated()]
duplicate_rows
duplicate_rows.count()
data = data.drop_duplicates()
duplicate_rows.count()
```

```python
x = data.iloc[:,:-1].values     
y = data.iloc[:,-1].values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)
```

### Model Building

Training the machine learning algorithms on the training data and making predictions on test data.

### Multiple Linear Regression

Multiple Linear Regression is a statistical method used to study the linear relationship between a dependent variable and multiple independent variables.

```python
from sklearn.linear_model import LinearRegression

lrg = LinearRegression()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
lrg.fit(X_train,y_train)
```

```python
y_pred_lrg = lrg.predict(X_test)
```

### Decision Tree
Decision Tree Regressor tries to predict a continuous target variable by cutting the feature variables into small zones, and each zone will have one prediction. A decision tree is a decision support hierarchical model that uses a tree-like model of decisions and their possible consequences, including chance event outcomes, resource costs, and utility.

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(max_depth=6)
```

```python
dtr.fit(X_train, y_train)
```

```python
y_pred_dtr = dtr.predict(X_test)
```

### Random Forest Regressor

Random Forest is a popular machine learning algorithm used for classification and regression tasks due to its high accuracy, robustness, feature importance, versatility, and scalability. Random Forest reduces overfitting by averaging multiple decision trees and is less sensitive to noise and outliers in the data. It provides a measure of feature importance, which can be useful for feature selection and data interpretation.It combines multiple decision tree regressors (a forest) to make more
accurate predictions. Each tree is built on a random subset of the data
(bagging) and random subsets of features (feature bagging).

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=100)
```

```python
rfr.fit(X_train, y_train)
```

```python
y_pred_rfr = rfr.predict(X_test)
```

### K-Nearest Neighbors (KNN)

K-nearest neighbors (KNN) is a supervised or unsupervised machinelearning algorithm used for both classification and regression tasks.KNN predicts the target variable of an instance by considering the majorityclass (for classification) or the mean (for regression) of its K-nearest neighbors in the feature space.

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
k = 62
knn = KNeighborsRegressor(n_neighbors=k)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print("Model\t\t\t RMSE \t\t MSE \t\t MAE \t\t R2")
print("""LinearRegression \t {:.2f} \t\t {:.2f} \t\t{:.2f} \t\t{:.2f}""".format(
            np.sqrt(mean_squared_error(y_test, y_pred_lrg)),mean_squared_error(y_test, y_pred_lrg),
            mean_absolute_error(y_test, y_pred_lrg), r2_score(y_test, y_pred_lrg)))
print("""DecissionRegression \t {:.2f} \t\t {:.2f} \t\t{:.2f} \t\t{:.2f}""".format(
            np.sqrt(mean_squared_error(y_test, y_pred_dtr)),mean_squared_error(y_test, y_pred_dtr),
            mean_absolute_error(y_test, y_pred_dtr), r2_score(y_test, y_pred_dtr)))
print("""RandomforestReg \t {:.2f} \t\t {:.2f} \t\t{:.2f} \t\t{:.2f}""".format(
            np.sqrt(mean_squared_error(y_test, y_pred_rfr)),mean_squared_error(y_test, y_pred_rfr),
            mean_absolute_error(y_test, y_pred_rfr), r2_score(y_test, y_pred_rfr)))
print("""K-NearestNeighbour\t {:.2f} \t\t {:.2f} \t\t{:.2f} \t\t{:.2f}""".format(
            np.sqrt(mean_squared_error(y_test, y_pred_knn)),mean_squared_error(y_test, y_pred_knn),
            mean_absolute_error(y_test, y_pred_knn), r2_score(y_test, y_pred_knn)))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns



fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize=(12,4))

ax1.scatter(y_pred_lrg, y_test, s=20)
ax1.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
ax1.set_ylabel("True")
ax1.set_xlabel("Predicted")
ax1.set_title("Linear Regression")

ax2.scatter(y_pred_dtr, y_test, s=20)
ax2.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
ax2.set_ylabel("True")
ax2.set_xlabel("Predicted")
ax2.set_title("Decission Regression")

ax3.scatter(y_pred_rfr, y_test, s=20)
ax3.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
ax3.set_ylabel("True")
ax3.set_xlabel("Predicted")
ax3.set_title("Random forest Regression")

ax4.scatter(y_pred_knn, y_test, s=20)
ax4.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
ax4.set_ylabel("True")
ax4.set_xlabel("Predicted")
ax4.set_title("K-Nearest Neighbour")

fig.suptitle("True vs Predicted")
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}'.format(height), xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

models = [lrg, dtr, rfr, knn]
names = ["Linear Regression", "Decision Tree Regressor", "Random Forest Regressor","K-Nearest Neighbour"]
rmses = []

for model in models:
    rmses.append(np.sqrt(mean_squared_error(y_test, model.predict(X_test))))

x = np.arange(len(names)) 
width = 0.3

fig, ax = plt.subplots(figsize=(10,7))
rects = ax.bar(x, rmses, width)
ax.set_ylabel('RMSE')
ax.set_xlabel('Models')
ax.set_title('RMSE with Different Algorithms')
ax.set_xticks(x)
ax.set_xticklabels(names, rotation=45)
autolabel(rects)
fig.tight_layout()
plt.show()
```

```python
req_col_names = ["MaterialQuantity", "AdditiveCatalyst", "AshComponent", "Water", "Plasticizer",
                 "ModerateAggregator", "RefinedAggregator", "FormulationDuration", "CompressionStrength"]


feature_dtr = dtr.feature_importances_
feature_rfr = rfr.feature_importances_

labels = req_col_names[:-1]
```

```python
x = np.arange(len(labels)) 
width = 0.3

fig, ax = plt.subplots(figsize=(10,6))
rects1 = ax.bar(x-(width/2), feature_dtr, width, label='Decision Tree')
rects2 = ax.bar(x+(width/2), feature_rfr, width, label='Random Forest')

ax.set_ylabel('Importance')
ax.set_xlabel('Features')
ax.set_title('Feature Importance')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend(loc="upper left", bbox_to_anchor=(1,1))

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}'.format(height), xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```
